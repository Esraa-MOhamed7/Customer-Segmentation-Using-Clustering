## Data Transformation

```python
la=LabelEncoder()
data["Gender"]=la.fit_transform(data["Gender"])
```

```python
data.dtypes
```

## Modeling

```python
num_of_clusters=[]
errors=[]
```

```python
for i in range(1,12):
    model=KMeans(n_clusters=i)
    model.fit(data)
    num_of_clusters.append(i)
    errors.append(model.inertia_)
```

```python
plt.plot(num_of_clusters, errors, marker='*')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (WCSS)")
plt.show()
```

```python
model=KMeans(n_clusters=4)
```

```python
model.fit(data)
```

```python
pred=model.predict(data)
```

```python
data["K_Means"]=pred
data
```

## Post-Clustering Analysis

```python
df1=data[data["K_Means"]==0]
df2=data[data["K_Means"]==1]
df3=data[data["K_Means"]==2]
df4=data[data["K_Means"]==3]
```

```python
plt.scatter(df1["Annual Income (k$)"],df1["Spending Score (1-100)"],label="Group1")
plt.scatter(df2["Annual Income (k$)"],df2["Spending Score (1-100)"],label="Group2")
plt.scatter(df3["Annual Income (k$)"],df3["Spending Score (1-100)"],label="Group3")
plt.scatter(df4["Annual Income (k$)"],df4["Spending Score (1-100)"],label="Group4")
plt.legend()
plt.show()
```

```python
data['K_Means'].value_counts().plot.pie(autopct="%0.2f%%")
plt.title("Customer Distribution by Cluster")
plt.show()
```

```python
cluster_means = data.groupby('K_Means').mean(numeric_only=True)
cluster_means.plot(kind='bar')
plt.title("Average Feature Values by Cluster")
plt.ylabel("Average Value")
plt.show()
```

```python
sns.boxplot(x=data["K_Means"], y=data["Spending Score (1-100)"])
plt.title("Spending Score by Cluster")
plt.show()
```

```python
sns.countplot(x="K_Means", hue="Gender", data=data)
plt.title("Cluster Distribution by Gender")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.show()
```

```python
summary = data.groupby("K_Means").agg({
    "Age": "mean",
    "Annual Income (k$)": "mean",
    "Spending Score (1-100)": "mean"
}).round(1)
```
